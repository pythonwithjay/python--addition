```python
#creating calculator using function
#creating function
#addition of two number
def addition(value1,value2):
        return value1+value2


#substraction of two number
def substract(value1,value2):
        return value1-value2


#multiplication of two number
def multiply(value1,value2):
        return value1*value2



#Divisinof two number
def divide(value1,value2):
        return value1/value2



#Average of two number
def average(value1,value2):
        return (value1+value2)/2

#now collect value fro user
print("please select a calulation:\n"
     "1. Addition\n"
     "2. Substraction\n"
     "3. Multiplication\n"
     "4. Division\n"
     "5. Average\n")
select=input("Enter a claculation fromm 1,2,3,4,5:")
number1=int(input("enter first value:"))
number2=int(input("Enter seond value :"))
#print the result
if select==1:
    print(number1,"+", number2, "=", addition(number1, number2))
elif select==2:
    print(number1,"-", number2, "=", substract(number1, number2))
elif select==3:
    print(number1,"*", number2, "=", multiply(number1, number2))
elif select==4:
    print(number1,"/", number2, "=", divide(number1, number2))
elif select==5:
    print("average of", number1,",", number2, "=", avergae(number1, number2))

else:
    print("invalid calculation")



```
